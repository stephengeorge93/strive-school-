```python
import pandas as pd
import numpy as np
import bs4 as bs
import requests
import csv


title =[]
link=[]
author=[]
ratings=[]
score=[]



page = 1

while page != 11:

    url = 'https://www.goodreads.com/list/show/6.Best_Books_of_the_20th_Century?page=' + str(++page)

    grURL = requests.get(url)
    grHTML = grURL.text

    soup = bs.BeautifulSoup(grHTML, 'html.parser')
    booklink = soup.find_all('a', attrs={'class': 'bookTitle'})
    authorlink =soup.find_all('a',attrs={'class':"authorName"})
    ratinglink=soup.find_all('span',attrs={'class':'minirating'})
    scorelink=soup.find_all('span',attrs={'class':'smallText uitext'})
   
    
    for links in booklink:
        bookname = links.get_text().replace('\n','')
        booklink = links['href']
      
        title.append(bookname)
        link.append(booklink)
       
    
    for links in authorlink:
        authorname=links.get_text()
        authorlink=links['href']
        author.append(authorname)
    
    for links in ratinglink:
        rating =links.get_text()
        ratings.append(rating)
    
    for links in scorelink:
        scoreno=links.get_text().strip().replace('\n','').replace('and','')
        score.append(scoreno)

    
    
   
    if page == 11:
        break
    page += 1
    print ('pageNo:' + str(page))
  
    
```

```python
data =list(zip(title,link,author,ratings,score))
df =pd.DataFrame(data,columns=['title','link','author','rating','score'])

    
        
        
    
```
